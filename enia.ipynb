```python
from scipy.stats import shapiro

# Prueba de normalidad para Control y Test
stat_c, p_c = shapiro(filtered_control_delta_t['log_total_delta_t_seconds'])
stat_t, p_t = shapiro(filtered_test_delta_t['log_total_delta_t_seconds'])

print(f"Control: p-value = {p_c}")
print(f"Test: p-value = {p_t}")

if p_c < 0.05:
    print("La distribución de Control no es normal.")
if p_t < 0.05:
    print("La distribución de Test no es normal.")
```

```python
#Dado que ninguna de las dos distribuciones es normal, probamos a hacer dos pruebas no parametricas

from scipy.stats import mannwhitneyu

# Mann-Whitney U Test
stat, p = mannwhitneyu(filtered_control_delta_t['log_total_delta_t_seconds'], filtered_test_delta_t['log_total_delta_t_seconds'])

print(f"Mann-Whitney U: Estadístico = {stat}, p-value = {p}")
if p < 0.05:
    print("Hay una diferencia significativa entre las distribuciones.")
else:
    print("No hay diferencia significativa entre las distribuciones.")
```

```python
from scipy.stats import ks_2samp

# Kolmogorov-Smirnov test
stat, p = ks_2samp(filtered_control_delta_t['log_total_delta_t_seconds'], filtered_test_delta_t['log_total_delta_t_seconds'])

print(f"KS Test: Estadístico = {stat}, p-value = {p}")
if p < 0.05:
    print("Control y Test tienen distribuciones significativamente diferentes.")
else:
    print("Control y Test no tienen distribuciones significativamente diferentes.")
```

```python
# Calcular retrocesos por paso en cada versión
def calculate_backtracks(df):
    # Mapear process_step a números
    step_order = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'step_4': 4, 'confirm': 5}
    df['step_order'] = df['process_step'].map(step_order)
    
    # Identificar retrocesos
    df['is_backtrack'] = df.groupby('visit_id')['step_order'].diff() < 0 #Si el usuario retrocede, el valor de la columna es True, si no False
    
    # Contar retrocesos por paso
    backtrack_counts = df[df['is_backtrack']].groupby('process_step').size().reset_index(name='backtrack_count') #df[df['is_backtrack']] selecciona solo las filas donde is_backtrack es True y .size() cuenta cuántas veces aparece cada paso en el conjunto filtrado.
    return backtrack_counts #.reset_index(name='backtrack_count') convierte el resultado en un DataFrame con dos columnas (process_step y backtrack_count)

# Obtener retrocesos para ambas versiones
control_backtracks = calculate_backtracks(df_control)
test_backtracks = calculate_backtracks(df_test)

# Agregar columna para identificar la versión
control_backtracks['version'] = 'Control'
test_backtracks['version'] = 'Test'

# Combinar los datos en un solo DataFrame
backtracks = pd.concat([control_backtracks, test_backtracks])

# Graficar el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=backtracks, x='process_step', y='backtrack_count', hue='version', palette=['blue', 'orange'])

# Títulos y etiquetas
plt.title('Cantidad de retrocesos por paso (Control vs Test)')
plt.xlabel('Paso del proceso (process_step)')
plt.ylabel('Cantidad de retrocesos')
plt.legend(title='Versión')
plt.tight_layout()
plt.show()


backtracks.to_csv("backtracks.csv")
```
